# What is Pytorch ?

Pytorch is an open-source Deep Learning librairie developped by Facebook's AI Research lab (FAIR). It's widely use to develop and train DL network. There are a keys points :

- Dynamic Tensor: Pytorch gives Tensor with automatic gradients compute. Tensors are dynamics, i-e it's possible to modify very easily the structure of the network.
- Autograd : Pytorch uses a system call Autograd for automatic gradients compute. It facilitate the implementation of backpropagation and to update weights for model during the training phase.
- High Level Modules : Pytorch provides high level API (Aplication Programming Interface) called `torch.nn`, which contains modules for build complex networks like CNN, RNN...
- GPU Support : Pytorch supports acceleration via GPU.
- Community and ecosystem : Pytorch has a large community and a vast ecosystem for librairies and tools, like torchvision pour images, torchaudio for son processing and pre-trained models available on hub like Hugging Face.

Pytorch is available [here](https://pytorch.org/get-started/locally/).
